```python
# Question: How does the pay gap change based on how many hours someone works?
# Import data on Median weekly earnings from shortest to longest workweek
medEarnByHrsWrkd = pd.read_csv(r'C:\Users\Student\Documents\DATS 6103- Individual Project 2- Iliana Maifeld-Carucci/medEarningsByHrsWrkd.csv', encoding= 'ISO-8859-1')
medEarnByHrsWrkd.head()
```

```python
#drop na values
medEarnByHrsWrkd.dropna(inplace=True)
medEarnByHrsWrkd
```

```python
#check that all variables are floats or integers
medEarnByHrsWrkd.dtypes
```

```python
#Delete rows 'Total, 16 years and older', '0 to 34 hours', '35 hours and over', '41 hours and over', 'Hours vary', 'Usually less than 35 hours', 'Usually 35 or more hours' so the hours are nicely discretized
medEarnByHrsWrkd.drop([0,1,9,12,17,18,19], inplace = True)
```

```python
medEarnByHrsWrkd
```

```python
#Create a plotly graph of median weekly pay for men and women as hours worked per week increases
x_g1 = medEarnByHrsWrkd.iloc[:, 0]
y1_g1 = medEarnByHrsWrkd.iloc[:, 5]
y2_g1 = medEarnByHrsWrkd.iloc[:, 8]
bubble1_g1 = medEarnByHrsWrkd.iloc[:, 4]
bubble2_g1 = medEarnByHrsWrkd.iloc[:, 7]
sizeref1_g1 = 2*max(bubble1_g1)/(70**2)
sizeref2_g1 = 2*max(bubble2_g1)/(70**2)

#create trace of women's pay
trace1_g1 = go.Scatter(x = x_g1, y = y1_g1, mode = 'markers', name = "Women's Earnings", marker = dict(symbol='circle',
        sizemode='area',
        sizeref=sizeref1_g1,
        size=medEarnByHrsWrkd.iloc[:, 4],
        line=dict(
            width=2
        ), sizemin=5))

#create trace of men's pay
trace2_g1 = go.Scatter(x = x_g1, y = y2_g1, mode = 'markers', name = "Men's Earnings", marker = dict(symbol='circle',
        sizemode='area',
        sizeref=sizeref2_g1,
        size=medEarnByHrsWrkd.iloc[:, 7],
        line=dict(
            width=2
        ), sizemin = 5))

data_g1 = [trace1_g1, trace2_g1]
layout_g1 = go.Layout(title = "Median Weekly Earnings By Hours Worked for Men and Women"+'<br />(Hover for value)', xaxis = dict(title = "Hours Worked",), yaxis = dict(title = "Weekly Earnings ($)",),)
fig_g1 = go.Figure(data = data_g1, layout = layout_g1)
py.iplot(fig_g1, filename = "medEarnByHrsWrkd") 
```
